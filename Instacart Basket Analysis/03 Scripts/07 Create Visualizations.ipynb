### Bar charts

```python
df['orders_day_of_week'].value_counts().sort_index()
```

```python
# create a bar chart with index order
bar = df['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['blue', 'blue', 'green', 'green', 'green', 'green', 'green'])

# Set labels for the x-axis with abbreviations
weekdays = ['Sat', 'Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri']

# Set x-axis labels and rotate them by 45 degrees
plt.xticks(range(len(weekdays)), weekdays, rotation=0)

# Add title and labels for axes
plt.title('Orders per Day of Week')
plt.ylabel('Number of Orders')
plt.xlabel('Day of Week')

# Display the plot
plt.show()
```

```python
# export bar chart as png
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

## 03 Histogram of the “order_hour_of_day” column

```python
#creating a histogram
hist_3 = df['order_hour_of_day'].plot.hist(bins = 24)

# Customize x-axis ticks
plt.xticks([0, 4, 8, 12, 16, 20])

# Highlight the time range between 9 and 17 hours
plt.axvspan(9, 17, color='yellow', alpha=0.3, label='9 AM to 5 PM')

# Add title and labels for axes
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')

```

#### the histogram demonstrates peak hours for order is between 9am - 5pm

```python
# saving the histogram
hist_3.figure.savefig(os.path.join(path, '04 Analysis/Visualizations/hist_peak_hours_of_the_day.png'))
```

## 04 Create a bar chart from the “loyalty_flag” column

```python
# create a bar chart of loyalty flag values
bar_loyalty = df['loyalty_flag'].value_counts().sort_index().plot.barh(ylabel="loyalty of customers", x="number of orders")

plt.title("Number of orders by loyality of customers", fontsize=16)
bar_loyalty.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda x,pos: format(x/1000000,'1.0f')+'M'))
```

#### Almost half of all orders is made by Regular customers. 

```python
# saving the bar
bar_loyalty.figure.savefig(os.path.join(path, '04 Analysis/Visualizations/bar_number_of_orders_by_loyality.png'))
```

#### family_status

```python
# extract fam_status and user_id into separate dataframe
df_family = df[['fam_status','user_id']]
```

```python
df_family.head()
```

```python
## Count unique users per fam_status
family_status = df_family.groupby(['fam_status'], as_index=False).agg({'user_id': 'nunique'})
```

```python
family_status
```

```python
# Create a bar graph to show the user distribution based on family status
bar_fam_status = sns.barplot(y='fam_status', x='user_id', data=family_status)
bar_fam_status.set(xlabel='Number of Users')
plt.title('Number of Customers by Family Status', fontsize=14)
fig = plt.figure(figsize=(20, 12))
```

```python
bar_fam_status.figure.savefig(os.path.join(path, '04 Analysis/Visualizations/bar_customer_by_fam_status.png'))
```

#### Most customers are married

## 07 Explore connection between age and income - visualize this relationship

```python
# Extract age and income into a separate dataframe
df_age_income = df[['age', 'income']]
```

```python
# checking the head
df_age_income.head()
```

```python
# Create a scatterplot 

#scatter = sns.scatterplot(x='age', y='income', data=df_age_income)
#plt.grid()
#plt.title('Do elder people have more spending power?')
#scatter.set(xlabel='age', ylabel='income')

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x='age', y='income', data=df_age_income, alpha=0.7, edgecolor='w', linewidth=0.5)

# Add grid for better visibility
plt.grid(True, linestyle='--', alpha=0.6)

plt.title('Relationship Between Age and Income')
scatter.set(xlabel='Age', ylabel='Income')

plt.show()
```

#### ages between 18 and 40 dont have more than 400 000 income while most are below 200 000
#### above 40 incomes erratic range up to 600 000 while most are below 300 000.


```python
scatter.figure.savefig(os.path.join(path, '04 Analysis/Visualizations/scatter_age_income.png'))
```
