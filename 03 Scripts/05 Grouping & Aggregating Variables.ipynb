## 03 Create loyalty flag for existing customers 
using the transform() and loc() functions


#### Transform function

```python
# creating an new column max_order and calculating the results
opm['max_order'] = opm.groupby(['user_id'])['order_number'].transform(np.max)
```

#### loc fuctions

```python
# Creating a red flag for customers with > 40 orders
opm.loc[opm['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Creating a red flag for customers with at least 11 orders and max 40
opm.loc[(opm['max_order'] <= 40) & (opm['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Creating a red flag for customers with 10 or less ordered products
opm.loc[opm['max_order'] <=10, 'loyalty_flag'] = 'New Customer'
```

## 05 Create spending_flag
The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

    If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Creating a new column mean_price
opm['mean_price']= opm.groupby(['user_id'])['prices'].transform('mean')
```

```python
opm.head()
```

```python
# Creating a new column 'spendings'
opm.loc[opm['mean_price'] < 10, 'spendings'] = 'Low spender'
```

```python
# Creating a new column 'spendings'
opm.loc[opm['mean_price'] >= 10, 'spendings'] = 'High spender'
```

```python
# Checking 'spendings' value_counts
opm['spendings'].value_counts()
```

## 06 Create order_frequency flag
In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Creating a column 'frequency median'
opm['frequency_median']=opm.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Checking dataframes head
opm.head()
```

```python
# Creating a new column order_frequency and assigning 'Non-frequent customer'

opm.loc[opm['frequency_median'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
# Creating a new column order_frequency and assigning 'Regular customer'
opm.loc[(opm['frequency_median'] <= 20) & (opm['frequency_median'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
# Creating a new column order_frequency and assigning 'Frequent customer'

opm.loc[opm['frequency_median'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Checking 'order_frequency's value_counts
opm['order_frequency'].value_counts()
```
